```python
# This one got like 4000+ ms time complexity


class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        count = 0
        prev = 0
        for element in nums:
            if prev != element:
                count = 0
            prev = element
            count += 1


            if element == prev and count > 2:
                nums[nums.index(element)] = None

        while None in nums:
            nums.remove(None)
        
        return len(nums)
    
nums = [0,0,1,1,1,1,2,3,3]
solution = Solution()
result = solution.removeDuplicates(nums)
print("Updated nums:", nums)
print("Length of updated nums:", result)
```

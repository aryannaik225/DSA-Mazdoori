```python
# Bro this cleared like 76 testcases, then some psychopathic input showed up and boom, memory exceeded.. like bruh, now have to design a new solution
# I am still so proud of myself to figure this one out myself.

class Solution(object):
  def canJump(self, nums):

    if len(nums)>0 and nums[0] == 0:
      return False

    n = len(nums)-1

    if n<=0:
      return True
    
    for i in range(n-1, -1, -1):
      if i+nums[i]>=n:
        return self.canJump(nums[:i+1])
    return False
        

nums = [2,3,1,1,4]
solution = Solution()
result = solution.canJump(nums)
print(result)

nums = [3,2,1,1,4]
result = solution.canJump(nums)
print(result)

nums = [3,2,1,0,4]
result = solution.canJump(nums)
print(result)
```

```python
# So the problem is that im creating new list in every recurssive window.. so have to design something that occurs within a single function.

class Solution(object):
  def canJump(self, nums):
    n = len(nums) - 1

    for i in range(len(nums)-2, -1, -1):
      if i + nums[i] >= n:
        n = i

    return n == 0
  
nums = [2,3,1,1,4]
solution = Solution()
result = solution.canJump(nums)
print(result)

nums = [3,2,1,1,4]
result = solution.canJump(nums)
print(result)

nums = [3,2,1,0,4]
result = solution.canJump(nums)
print(result)


# Succeeded
```

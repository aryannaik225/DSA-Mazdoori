```python
class Solution(object):
    def removeDuplicates(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        
        arr = []
        i = 0
        while i < len(nums):
            if nums[i] in arr:
                nums.pop(i)
            else:
                arr.append(nums[i])
                i += 1
                
        return len(arr)
    
if __name__ == '__main__':
    nums = [1,1,2]
    print(Solution().removeDuplicates(nums))
    print(nums)
    nums = [0,0,1,1,1,2,2,3,3,4]
    print(Solution().removeDuplicates(nums))
    print(nums)
    nums = [1,1]
    print(Solution().removeDuplicates(nums))
    print(nums)
```

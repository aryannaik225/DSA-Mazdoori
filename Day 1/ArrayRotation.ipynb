```python
# Complete the 'rotateLeft' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER d
#  2. INTEGER_ARRAY arr
#

# HackerRank
def rotateLeft(d, arr):
    # Write your code here
    for i in range(d):
        arr.append(arr.pop(0))
    return arr

# LeetCode
def rotateRight(d, arr1):
    # Write your code here
    for i in range(d):
        arr1.insert(0, arr1.pop(-1))
    return arr1


if __name__ == '__main__':
  n = 5
  d = 4
  arr = [1, 2, 3, 4, 5]

  result = rotateLeft(d, arr)
  print(' '.join(map(str, result)))

  result1 = rotateRight(d, arr)
  print(' '.join(map(str, result1)))
```

### Intuition:
The problem requires rotating the array to the right by ( k ) steps. This can be achieved efficiently by using array manipulation instead of performing ( k ) rotations iteratively.

### Approach:
1. **Modulo Operation**: Since rotating ( n ) times brings the array back to its original state, ( k ) can be reduced to ( k \mod n ).
2. **Reverse Technique**:
  - Reverse the entire array.
  - Reverse the first ( k ) elements.
  - Reverse the remaining ( n-k ) elements.
3. The final array will be the desired rotated version.

### Complexity:
- **Time Complexity**: ( O(n) ), where ( n ) is the length of the array, as we perform three passes over the array.
- **Space Complexity**: ( O(1) ), as the rotation is performed in-place.

```python
def rotateRight(k, nums):
  n = len(nums)
  k %= n
  nums.reverse()
  nums[:k] = reversed(nums[:k])
  nums[k:] = reversed(nums[k:])
  return nums

if __name__ == '__main__':
  nums = [1, 2, 3, 4, 5, 6, 7]
  k = 3

  result = rotateRight(k, nums)
  print(' '.join(map(str, result)))
```

```python
# or this way

def rotateRight(k, nums):
  k = k % len(nums)
  if k == 0:
    return nums
  nums[:] = [nums[i-k] for i in range(len(nums))]
  return nums

if __name__ == '__main__':
  nums = [1, 2, 3, 4, 5, 6, 7]
  k = 3

  result = rotateRight(k, nums)
  print(' '.join(map(str, result)))
```
